```python
import statsmodels.api as sm
from sklearn.model_selection import train_test_split

# Iteratively choosing predictors with test_size= 0.25 and random_state = 1
# based on correlations with target attribute seen above
predictor = data.filter(['Log GDP Per Capita']) # r-sq: 0.2938, rmse: 0.9103
predictor = data.filter(['Social Support']) # r-sq: 0.4301, rmse: 0.8177
predictor = data.filter(['Life Expectancy']) # r-sq: 0.317, rmse: 0.8952
predictor = data.filter(['Freedom of Choice']) # r-sq: 0.1943, rmse: 0.9723
predictor = data.filter(['Generosity']) # r-sq: 0.0598, rmse: 1.0504          OUT
predictor = data.filter(['Corruption']) # r-sq: 0.097, rmse: 1.0294           OUT
predictor = data.filter(['Delivery Quality']) # r-sq: 0.4621, rmse: 0.7945
predictor = data.filter(['Democratic Quality']) # r-sq: 0.3304, rmse: 0.8864
predictor = data.filter(['Positive Affect']) # r-sq: 0.2234, rmse: 0.9546


predictor = data.filter(['Social Support', 'Delivery Quality']) # r-sq: 0.5935, rmse: 0.6906
predictor = data.filter(['Social Support', 'Life Expectancy']) # r-sq: 0.5105, rmse: 0.7579
predictor = data.filter(['Delivery Quality', 'Life Expectancy']) # r-sq: 0.5214, rmse: 0.7494
predictor = data.filter(['Social Support', 'Delivery Quality', 'Life Expectancy']) # r-sq: 0.6209, rmse: 0.6669

predictor = data.filter(['Social Support', 'Delivery Quality', 'Democratic Quality']) # r-sq: 0.5988, rmse: 0.6862
predictor = data.filter(['Social Support', 'Delivery Quality', 'Life Expectancy', 'Democratic Quality']) # r-sq: 0.6238, rmse: 0.6644


predictor = data.filter(['Social Support', 'Delivery Quality', 'Life Expectancy', 'Democratic Quality', 'Log GDP Per Capita']) # r-sq: 0.624, rmse: 0.6642 NOT ENOUGH DIFF
predictor = data.filter(['Social Support', 'Delivery Quality', 'Life Expectancy', 'Democratic Quality', 'Positive Affect']) # r-sq: 0.6415, rmse: 0.6486
predictor = data.filter(['Social Support', 'Delivery Quality', 'Life Expectancy', 'Democratic Quality', 'Freedom of Choice']) # r-sq: 0.6399, rmse: 0.65
predictor = data.filter(['Social Support', 'Delivery Quality', 'Life Expectancy', 'Democratic Quality', 'Positive Affect', 'Log GDP Per Capita']) # r-sq: 0.6417, rmse: 0.6484  NOT ENOUGH DIFF
predictor = data.filter(['Social Support', 'Delivery Quality', 'Life Expectancy', 'Democratic Quality', 'Positive Affect', 'Freedom of Choice']) # r-sq: 0.6463, rmse: 0.6442 NOT ENOUGH DIFF
#highest r-sq, lowest rmse but likely overfit
predictor = data.filter(['Year','Log GDP Per Capita','Social Support','Life Expectancy','Freedom of Choice','Generosity','Corruption','Delivery Quality','Democratic Quality']) # r-sq: 0.6706, rmse: 0.6217 


# CHOSEN PREDICTORS
predictor = data.filter(['Social Support', 'Delivery Quality', 'Life Expectancy', 'Democratic Quality', 'Positive Affect']) # r-sq: 0.6415, rmse: 0.6486
target = data.filter(['Life Ladder'])

X = predictor
y = target['Life Ladder']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,random_state=1)
model = sm.OLS(y_train, X_train).fit()
predictions = model.predict(X_test)
model.summary()
```
