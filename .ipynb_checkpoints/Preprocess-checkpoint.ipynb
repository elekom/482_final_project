```python
import pandas as pd

# dataset with happiness index and other data
columns = ["Country", "Year", "Life Ladder", "Log GDP Per Capita", "Social Support", "Life Expectancy", "Freedom of Choice", "Generosity", "Corruption", "Positive Affect", "Negative Affect", "Confidence in National Government", "Democratic Quality", "Delivery Quality", "d1", "d2", "d3", "d4", "Gini of Household Income"]
data = pd.read_csv("HappinessIndex+.csv", names = columns, header = 0)
data.drop(columns = ["d1", "d2", "d3", "d4"], inplace = True)
data
```

```python
# dataset with population and population density
columns = ["Country", "Year", "Population", "Population Density", "d1", "d2", "d3", "d4", "d5", "d6", "d7", "d8", "d9", "d10", "d11"]
dataPop = pd.read_csv("migration_population.csv", names = columns, header=0)
dataPop = dataPop.filter(["Country", "Year", "Population", "Population Density"])
data["Year"] = data["Year"].astype('object')
dataPop
```

```python
# Preprocessed dataset
data = pd.merge(left=data, right=dataPop, how='left', left_on=['Country','Year'], right_on=['Country', 'Year'])
dataIndex = data.filter(["Country","Year","Life Ladder"])
data
```

```python
import copy
import numpy as np

def MF(matrix, k, maxiter):
    U = np.ones((matrix.shape[0], k), dtype=np.int32)
    M = np.ones((matrix.shape[1], k), dtype=np.int32)
    
    mat = copy.deepcopy(matrix)

    for i in range(maxiter):
        
        U = np.multiply(U, np.dot(mat,M)/np.dot(np.dot(U,M.T),M))
        M = np.multiply(M, np.dot(mat.T,U)/np.dot(np.dot(M,U.T),U))
        UMT = np.dot(U,M.T)
        
        for i in range(matrix.shape[0]):
            for j in range(matrix.shape[1]):
                if matrix[i][j] == 0:
                    mat[i][j] = UMT[i][j]
             
    return mat
```

```python
# Apply the model to the ratings data. Set k=2 and maxiter = 100
dataFilled = MF(data.drop(columns=["Country","Year"]).fillna(0).values, 2, 500)
cols = ["Life Ladder", "Log GDP Per Capita", "Social Support", "Life Expectancy", "Freedom of Choice", "Generosity", "Corruption", "Positive Affect", "Negative Affect", "Confidence in National Government", "Democratic Quality", "Delivery Quality", "Gini of Household Income","Population","Population Density"]

dataFilled = pd.DataFrame(dataFilled, columns=cols)
data.update(dataFilled)

# created preprocessed csv file
data.to_csv("dataComplete.csv", index = False)
data
```
